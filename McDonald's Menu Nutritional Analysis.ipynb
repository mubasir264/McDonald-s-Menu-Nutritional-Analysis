<h2 style="color: #a64d79;"><b>Extracting meaningful information from McDonald's menu nutritional dataset</b></h2>

<p><b><span style="color: #ff0000;">McDonald's Dataset</span></b> which contains <b><span style="color:#9900ff;">260 menu items</span></b>, organized into <b><span style="color:#9900ff;">9 distinct food categories</span></b>. This data captures the nutritional values of the items, focusing on five key nutrients: <b><span style="color:#bf9000;">Calories, Protein, Carbohydrates, Sodium, and Dietary Fiber</span></b>, and excludes certain nutrients and daily value percentages for a more streamlined analysis. The serving sizes in the dataset are huge but can be understood in two types: <b><span style="color: #990000;">Weight-based servings</span></b> and <b><span style="color: #990000;">Volume-based servings</span></b>.
</p>


<h2 style="color: #339966;">Food Categories</h2>

<ul>
  <li><b style="color: #FF9900;">Breakfast:</b> Breakfast items are rich in calories and protein but can also contain high levels of sodium and total fat, which are key to analyzing their overall health impact.</li>
  <li><b style="color: #FF9900;">Beef & Pork:</b> Beef and pork menu items are high in protein, but they also have significant amounts of cholesterol and total fat, making these nutrients critical in evaluating their nutritional value.</li>
  <li><b style="color: #FF9900;">Chicken & Fish:</b> These are excellent sources of protein while typically having lower fat levels than beef, but sodium and cholesterol content should be considered.</li>
  <li><b style="color: #FF9900;">Salads:</b> Salads are typically lower in calories and rich in vitamins A and C, calcium, and dietary fiber, offering a lighter, nutrient-dense option on the menu.</li>
  <li><b style="color: #FF9900;">Snacks & Sides:</b> These items often have high carbohydrate and sodium content, with some sides containing sugars that impact their nutritional assessment.</li>
  <li><b style="color: #FF9900;">Desserts:</b> Desserts are generally high in sugars, calories, and fat, which are key nutrients to monitor when analyzing their effects on overall diet.</li>
  <li><b style="color: #FF9900;">Beverages:</b> Many beverages contain high levels of sugars and calories, with some also contributing to daily sodium intake.</li>
  <li><b style="color: #FF9900;">Coffee & Tea:</b> While low in calories, coffee and tea drinks can vary greatly in sugar and fat content depending on the added ingredients like cream and syrups.</li>
  <li><b style="color: #FF9900;">Smoothies & Shakes:</b> Smoothies and shakes are typically high in sugars, total fat, and calories, making them a key focus when assessing indulgent menu opions.</li>
</ul>


<h2 style="color: #339966;">Key Nutrients in McDonald's Menu Items</h2>

<ul>
  <li><b style="color: #FF9900;">Calories:</b> Measure of energy provided by the food.</li>
  <li><b style="color: #FF9900;">Calories from Fat:</b> Portion of total calories that come from fat content.</li>
  <li><b style="color: #FF9900;">Total Fat:</b> The sum of all fats in the item.</li>
  <li><b style="color: #FF9900;">Total Fat (% Daily Value):</b> Percentage of daily recommended fat intake in one serving.</li>
  <li><b style="color: #FF9900;">Saturated Fat:</b> Type of fat that can raise cholesterol levels.</li>
  <li><b style="color: #FF9900;">Saturated Fat (% Daily Value):</b> Contribution to the daily recommended limit for saturated fats.</li>
  <li><b style="color: #FF9900;">Trans Fat:</b> Unhealthy fat that can increase the risk of heart disease.</li>
  <li><b style="color: #FF9900;">Cholesterol:</b> A substance linked to heart health, present in animal products.</li>
  <li><b style="color: #FF9900;">Cholesterol (% Daily Value):</b> How much of the daily recommended cholesterol the item contains.</li>
  <li><b style="color: #FF9900;">Sodium:</b> Salt content, which can affect blood pressure.</li>
  <li><b style="color: #FF9900;">Sodium (% Daily Value):</b> The percentage of the recommended daily salt intake in one serving.</li>
  <li><b style="color: #FF9900;">Carbohydrates:</b> Main source of energy from sugars, starches, and fibers.</li>
  <li><b style="color: #FF9900;">Carbohydrates (% Daily Value):</b> Contribution to the daily recommended intake of carbs.</li>
  <li><b style="color: #FF9900;">Dietary Fiber:</b> Aids digestion and promotes a feeling of fullness.</li>
  <li><b style="color: #FF9900;">Sugar:</b> Simple carbohydrates, can affect blood sugar levels.</li>
  <li><b style="color: #FF9900;">Protein:</b> Essential for muscle building and repair.</li>
  <li><b style="color: #FF9900;">Vitamin A (% Daily Value):</b> Contribution to daily vitamin A intake, important for vision and immune function.</li>
  <li><b style="color: #FF9900;">Vitamin C (% Daily Value):</b> Supports immune health and skin repair.</li>
  <li><b style="color: #FF9900;">Calcium (% Daily Value):</b> Essential for bone strength and health.</li>
  <li><b style="color: #FF9900;">Iron (% Daily Value):</b> Supports oxygen transport n the blood.</li>
</ul>

<h3 style="color: #A52A2A;">Identifying Outlilers in the data</h3>

```python
col=df.select_dtypes(exclude='object').columns
Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1
Outliers = df[~((df[col] < (Q1 - 1.5*IQR))| (df[col] > (Q3 + 1.5*IQR))).any(axis=1)]

len(Outliers)
```

***Due to the presence of 154 outliers in the data, it is preferable to use the median (central point) instead of the mean to obtain a clearer representation of the dataset.***

<h3><font color="#008080">4.2.3 Analyzing Nutritional Content Distribution through Boxplot by Food Categories</font></h3>

```python
nutritional_values = [
    'Calories', 'Carbohydrates', 'Dietary Fiber', 'Protein', 
    'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 
    'Calcium (% Daily Value)', 'Sodium', 'Iron (% Daily Value)', 
    'Total Fat', 'Cholesterol', 'Sugars'
]

palette = sns.color_palette("Set3", len(df['Category'].unique()))

for value in nutritional_values:
    plt.figure(figsize=(15, 8))
    sns.boxplot(x='Category', y=value, data=df, palette=palette, hue='Category', legend=False)
    plt.title(f'{value} by Category', fontsize=14, fontweight='bold')
    plt.xlabel('Category')
    plt.ylabel(value)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()  
    plt.savefig(f"{value}_boxplot.jpg", bbox_inches='tight') 
    plt.show()
    plt.close()
```

<p style="color: #333333;">Conducting a detailed nutritional analysis can significantly benefit McDonald's customers in several ways:</p>

  <li><b style="color: #FF4500;">Customization:</b> The analysis highlights specific menu items that may contain excessive fat, sodium, or sugar. This can guide customers who need to manage their intake of these nutrients due to medical conditions, such as hypertension or diabetes, enabling them to customize their meals to suit their health requirements.</li>

  <li><b style="color: #FF4500;">Dietary Preferences:</b> Customers can easily find menu options that cater to specific dietary preferences or restrictions. For instance, identifying high-protein items like Chicken & Fish or low-calorie beverages can help those focused on weight management or muscle gain.</li>
</ul>
